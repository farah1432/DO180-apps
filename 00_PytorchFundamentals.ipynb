## Matrix multiplication
Two ways of performing multiplication in neural networks and deep learning:
1. Element-wise multiplication
2. Matrix multiplication (dot product) most common
Most information on multiplying matrixs - https://www.mathsisfun.com/algebra/matrix-multiplying.html

There are two main rules that need to satisfy for matrix multiplication.

1. The ** inner dimensions ** must match:

(3, 2) @ (3, 2) won't work
(2, 3) @ (3, 2) will work
(3, 2) @ (2, 3) will work

2. The resulting matrix has the shape of the ** outer dimension **

(2, 3) @ (3, 2) shape (3, 2)


```python
# Element wise multiplication
print(tensor, "*" , tensor)
print(f"Equals: {tensor * tensor}")
```

```python
# Matrix multiplication
torch.matmul(tensor,tensor  )
```

```python
value = 0
for i in range(len(tensor)) :
  value += tensor[i] * tensor [i]
print(value)
```

## Reshaping, stacking, squeezing and unsqueezing
Reshaping - Reshapes an input tensor to a defined ShapeType
View - Return a view of an input tensor of certain shape but keep the same memory as the original tensor
Stacking - combine multiple tensors on top of each other (vstack or hstack)
Squeeze - removes all '1' dimensions from a tensor
Unsqueeze - add a '1' dimension to a target tensor
Permute - Return a view of the input with dimensions permuted (swapped) in a certain way

```python
## Create a tensor
import torch
x = torch.arange (1., 10.)
x, x.shape
```

```python
# Add an extra dimension (dimension must be multiply/compatible with the size)

X_reshaped = x.reshape(1, 9)
X_reshaped, X_reshaped.shape
```

```python
X_reshaped = x.reshape(9, 1)
X_reshaped, X_reshaped.shape
```

```python
# Change the view
z=x.view(1, 9)
z, z.shape
```

```python
# Changing z changes x (because the view of a tensor shares the same memory as the original)
z[ :, 7] =9
z,x
```

```python
# Stack tensors on top each other
x_stacked = torch.stack([x, x, x], dim=0)
x_stacked
```

```python
# torch.squeeze() - removes extra dimension from x_reshaped
print(f"Previous tensor: {X_reshaped}")
print(f"Previous shape: {X_reshaped.shape}")
# Remove extra dimensions from x_reshaped
x_squeezed = X_reshaped.squeeze()
print(f"New tensor: {x_squeezed}")
print(f"New shape: {x_squeezed.shape}")
```

```python
# torch.unsqueeze () - adds a single dimension to a target tensor at a specific dimension.
print(f"Previous target: {x_squeezed}")
print(f"Previous shape: {x_squeezed.shape}")

# add an extra dimension with unsqueeze
x_unsqueezed = x_squeezed.unsqueeze (dim=0)
print(f"New tensor: {x_unsqueezed}")
print(f"New shape: {x_unsqueezed.shape}")
```

```python
# torch.permute - rearranges the dimensions of a target tensor in a specified order
x_original = torch.rand(size=(224, 224, 3)) # Hieght, width, color channel
# permute the original tensor to rearrange the axis (or dim) order
x_permuted = x_original.permute(2, 0, 1) # shifts axis 0->1, 1->2, 2->0
print(f"Previous shape: {x_original.shape}")
print(f"New shape: {x_permuted.shape}")

# premute and original share the same space both are view
```

```python
x_original[0, 0, 0] = 728218
x_original[0, 0, 0], x_permuted[0, 0, 0]
```

## Reproducibility(trying to take random out of random)
In short how a nueral network learns:
start with random numbers -> tensor operations -> update random numbers to try and make them better representations of the data -> again -> again -> again...

To reduce the randomness in neural networks and PyTorch comes the concept of a **Random seed**

Essentially what the random seed does is "flavour" the randomness.

https://docs.pytorch.org/docs/stable/notes/randomness.html

```python
 ## Create two random tensors
 random_tensor_A = torch.rand(3, 4)
 random_tensor_B = torch.rand(3, 4)

 print(random_tensor_A)
 print(random_tensor_B)
 print(random_tensor_A == random_tensor_B)
```

```python
## Let's make some random but reproducible tensors
import torch

# Set the random seed
RANDOM_SEED = 42
torch.manual_seed(RANDOM_SEED) ## this code only work for 1 line
random_tensor_C = torch.rand(3, 4)
torch.manual_seed(RANDOM_SEED)
random_tensor_D = torch.rand(3, 4)
print (random_tensor_C)
print (random_tensor_D)
print (random_tensor_C == random_tensor_D)
```
